Task 1: Exploring and Visualizing a Simple Dataset
Dataset:
Iris Dataset

Problem Statement:

Understanding data is the first step in any AI/ML workflow. Raw datasets often contain hidden patterns, distributions, and relationships that must be explored before building models.

This task focuses on exploring the Iris dataset to identify feature relationships, value distributions, and potential outliers using statistical summaries and visualizations.

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 1: Load Dataset
df = pd.read_csv("Iris.csv")
```

```python
# Step 2: Inspect Dataset Structure
# Display shape of dataset
print("Shape (rows, columns):", df.shape)
# Display column names
print("\nColumn names:")
print(df.columns)
# Display first 5 rows
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Step 3: Summary Statistics
print("Info:")
df.info()
```

```python
# Statistical summary of numerical columns
print("Describe:")
df.describe()
```

Visualizing the dataset 

```python
# Step 4: Data Visualization
# Scatter Plot  Relationship between features
plt.figure(figsize=(6, 4))

sns.scatterplot(data=df,x="SepalLengthCm", y="SepalWidthCm", hue="Species")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.show()
```

```python
# Histogram  Distribution of values
plt.figure(figsize=(6, 4))
sns.histplot(df["PetalLengthCm"], bins=20, kde=True)
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()

df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].hist(figsize=(10, 8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
#Box Plot  Detect outliers
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Species",y="PetalLengthCm")
plt.title("Petal Length by Species")

plt.show()
```
